```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#data taken from: https://datahub.io/core/global-temp#resource-annual
data = pd.read_csv('globalData.csv')
data.shape
```

```python
data.plot("Year", "Mean")
```

```python
data.plot.scatter("Year", "Mean")
```

```python
#x_train, x_test, y_train, y_test = train_test_split(data.Year, data.Mean, test_size = 0.2)
x_train = data.iloc[:, 0].values.reshape(-1, 1)
y_train = data.iloc[:, 1].values.reshape(-1, 1)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
y_pred = regr.predict(x_train)
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, y_pred, color='red')
plt.show
```
